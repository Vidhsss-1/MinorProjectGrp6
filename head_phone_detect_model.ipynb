```python

import math
# Extract coordinates of the two detected objects
with open('/content/yolov5/runs/detect/exp/labels/teataryann.txt', 'r') as file:
    lines = file.readlines()
    objects = []
    for line in lines:
        data = line.split()
        if data[0] == '0' or data[0] == '1':
            objects.append(data[1:])

# Calculate the distance between the head and phone
head_coords = objects[0]
phone_coords = objects[1]

x1, y1 = float(head_coords[1]), float(head_coords[2])
x2, y2 = float(phone_coords[1]), float(phone_coords[2])

distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Print the distance
print(f"Distance between head and phone: {distance}")
```
